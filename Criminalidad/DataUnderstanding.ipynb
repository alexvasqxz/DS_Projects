### Columnas Numericas
Observando los valores Minimos y Maximo, asi como la media por columna notamos que:

- Para el Año de Inicio todo parece estar bien, teniendo un rango de 2019 a 2022

- El año de los hechos presenta un outlier con un valor de 222 que debe eliminarse, fuera de eso el promedio de 2020 tiene sentido

```python
# Minimos y Maximos de Columnas
import pandas as pd

data = pd.read_csv("sourceData.csv")
for col in data.columns:
    if(data[col].dtype in ['int64', 'float64']):
        print(col + ' (Min: ' + str(data[col].min()) + ' Max: ' + str(data[col].max()) + ')')

mean = data.mean(numeric_only=True)
print(mean)
```
